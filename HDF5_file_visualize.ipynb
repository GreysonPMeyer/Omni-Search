```python
import h5py
import matplotlib.pyplot as plt
import random

# Path to the HDF5 file
i=2
hdf5_file_path = "resized_images/resized_images_chunk_{}.h5".format(i)

# Function to visualize random images and metadata
def visualize_random_images(hdf5_file_path, num_images_to_display):
    with h5py.File(hdf5_file_path, 'r') as hdf5_file:
        # Get the number of images
        num_images = len(hdf5_file['images'])
        
        # Generate random indices
        random_indices = random.sample(range(num_images), num_images_to_display)
        
        for i in random_indices:
            # Extract image data
            image_data = hdf5_file[f'images/{i}'][:]
            
            # Extract metadata
            artist_full_name = hdf5_file[f'metadata/{i}/artist_full_name'][()].decode('utf-8')
            artwork_name = hdf5_file[f'metadata/{i}/artwork_name'][()].decode('utf-8')
            artwork_type = hdf5_file[f'metadata/{i}/artwork_type'][()].decode('utf-8')
            
            # Display the image
            plt.imshow(image_data)
            plt.title(f"Artist: {artist_full_name}\nArtwork: {artwork_name}\nType: {artwork_type}")
            plt.axis('off')
            plt.show()
            
            print("\n")

# Visualize 5 random images and metadata
visualize_random_images(hdf5_file_path, num_images_to_display=5)
```
